```python
#convertir el biciMad en un df para poder luego añadir las lon, lat y de ahí poder limpiar.
df_biciMad = pd.DataFrame(biciMad)
df_biciMad
```

```python
#separar las geometry coordinates por latitud y longitud para poder compararlas con las de los parques y jardines
lat = [row["geometry_coordinates"].split(",")[0].replace("[", "") for index, row in df_biciMad.iterrows()]
lat
```

```python
lon = [row["geometry_coordinates"].split(",")[1].replace("]", "") for index, row in df_biciMad.iterrows()]
lon
```

```python
#añadir longitud y latitud al df
df_biciMad ['latitud'] = lat
df_biciMad ['longitud'] = lon
df_biciMad
```

```python
#eliminar, "limpiar" aquellos datos que no me van a ser útiles en mi analisis. 
df_biciMad = df_biciMad.drop(['id', 'light', 'number', 'activate','geometry_type'], axis=1)
df_biciMad
```

```python
def to_mercator(lat, long):
    # transform latitude/longitude data in degrees to pseudo-mercator coordinates in metres
    c = gpd.GeoSeries([Point(lat, long)], crs=4326)
    c = c.to_crs(3857)
    return c
```
